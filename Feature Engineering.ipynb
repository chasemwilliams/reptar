```python
#create ID columns: race_id, horse_id, trainer_id, jockey_id, track_id
def create_id_col(df, new_id_column, old_id_column):
    df[new_id_column]=df[old_id_column].str.extract(r'(?:\()([^)]+)(?:\))')
```
